## Part 1: The Expense Data Model

We'll use a `dataclass` to represent an expense. This section is complete as a reference.

```python
@dataclass
class Expense:
    date: str          # Format: 'YYYY-MM-DD'
    category: str      # e.g., 'Food', 'Travel'
    amount: float      # Dollar amount
    description: str   # Brief description

    @staticmethod
    def validate_date(date_str: str) -> str:
        """Validate and normalize the date string."""
        try:
            dt = datetime.strptime(date_str.strip(), DATE_FMT)
            return dt.strftime(DATE_FMT)
        except ValueError:
            raise ValueError("Invalid date format. Please use YYYY-MM-DD.")

# Test it out!
test_expense = Expense(
    date="2025-01-15",
    category="Food",
    amount=12.50,
    description="Lunch at cafe"
)
print(f"Test expense created: {test_expense}")
```
