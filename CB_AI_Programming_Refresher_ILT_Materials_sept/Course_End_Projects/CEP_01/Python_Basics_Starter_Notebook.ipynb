## 1) Data Types and Data Assignment
**Goal:** Learn how to store values in variables and see common data types.


```python

# Objective: Data Types & Data Assignment
# TIP: Think of a variable as a "labeled box" where you store a value.

# Assigning (storing) values to variables
name = '01.9'          # string (text)
age = 35               # integer (whole number)
height = 5.11          # float (decimal number)
is_student = True      # boolean (True/False)

# Lists store multiple items in order (like a shopping list)
favorite_numbers = [7, 11, '42', True]
print(favorite_numbers)
# Dictionaries store key-value pairs (like a small phone book)
person = {
    "name": name,
    "age": age,
    "is_student": is_student
}

# Let's print them to see what's inside
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is a student:", is_student)
print("Favorite numbers:", favorite_numbers)
print("Person dictionary:", person)

# TIP: Use clear variable names (e.g., 'height_in_feet') so your code reads like English.
```


## 2) Operators in Python
**Goal:** Use math and comparison operators to do work with numbers and values.


```python

# Objective: Operators in Python
# TIP: Start with simple math, then try comparisons (>, <, ==) to ask yes/no questions.

apples = 3
bananas = 2

# Arithmetic operators
total_fruit = apples + bananas        # addition
difference = apples - bananas         # subtraction
product = apples * bananas            # multiplication
average = (apples + bananas) / 2      # division

print("Total fruit:", total_fruit)
print("Difference:", difference)
print("Product:", product)
print("Average:", average)

# Comparison operators return True or False
has_more_apples = apples > bananas    # Is 3 greater than 2?
equal_amount = apples == bananas      # Is 3 equal to 2?

print("Do we have more apples than bananas?", has_more_apples)
print("Do we have the same amount?", equal_amount)

# Logical operators: combine True/False values
is_small_and_equal = (total_fruit < 10) and (apples == bananas)
print("Less than 10 AND equal amounts?", is_small_and_equal)

# TIP: Use parentheses to make expressions easy to read.
```


## 3) Strings in Python
**Goal:** Work with text â€” cleaning it up, combining it, and formatting messages.


```python

# Objective: Strings in Python
# TIP: Strings are text in quotes. Use helpful tools like .strip(), .lower(), and .title().

raw_name = "   norm   "
clean_name = raw_name.strip().title()   # remove spaces and Title-Case it
greeting = f"Hello, {clean_name}! Welcome to Python."  # f-string formats text

print("Raw:", repr(raw_name))
print("Cleaned:", clean_name)
print(greeting)

# Joining small pieces into one string
words = ["learning", "python", "is", "fun"]
sentence = " ".join(words)
print("Joined sentence:", sentence)

# TIP: Use f-strings for readable messages: f"Your score is {score}"
```


## 4) File Handling in Python
**Goal:** Save information to a file and read it back later.


```python

# Objective: File Handling in Python
# TIP: Use a 'with' block to open files so Python closes them automatically.

from pathlib import Path

file_path = Path("hello_python.txt")

# Write text to a file (this will create the file if it doesn't exist)
with file_path.open("w", encoding="utf-8") as f:
    f.write("Hello from Python!\n")
    f.write("This is your first saved file.\n")

# Read the text back
with file_path.open("r", encoding="utf-8") as f:
    content = f.read()

print("File content:")
print(content)

# TIP: If you see a FileNotFoundError, check your file name and location.
```


## 5) Error Handling in Python
**Goal:** Prevent your program from crashing when something goes wrong.


```python

# Objective: Error Handling in Python
# TIP: Use try/except to "catch" problems and show a friendly message instead of crashing.

user_input = "abc"  # Imagine the user typed letters by mistake

try:
    number = int(user_input)  # This will fail for "abc"
    print("Nice! You typed the number:", number)
except ValueError:
    print("Oops! That wasn't a number. Please type digits only (e.g., 123).")

# TIP: Catch specific errors (like ValueError) so your message is accurate.
```
